# Metadata integration across all datasets

This notebook assembles all metadata for Supplementary Table on sample information from the MERFISH datasets generated in this project (coronal sections for aging, sagittal sections for aging, coronal sections for exercise, coronal sections for partial reprogramming, coronal sections from 140-gene pilot)

```python
import scanpy as sc
import pandas as pd
import os
import numpy as np

os.chdir("/labs/abrunet1/Eric/MERFISH")
```

```python
cols = ["mouse_id", "age", "slide_id", "cohort"]
rows = []
```

```python
# Coronal aging
adata = sc.read_h5ad("data/integrated_aging_coronal_celltyped_regioned_raw.h5ad")
adata = adata[(adata.obs.clusters!="1")&(adata.obs.mouse_id!="89")&(adata.obs.mouse_id!="67")]
```

```python
for mid in np.unique(adata.obs.mouse_id):
    sub_adata = adata[adata.obs.mouse_id==mid]
    row = []
    for col in cols:
        row.append(sub_adata.obs[col].values[0])
    row.append(sub_adata.shape[1])
    row.append(sub_adata.shape[0])
    rows.append(row)
```

```python
# Sagittal aging
adata = sc.read_h5ad("data/integrated_aging_sagittal_clustered_registered.h5ad")
```

```python
for mid in np.unique(adata.obs.mouse_id):
    sub_adata = adata[adata.obs.mouse_id==mid]
    row = []
    for col in cols:
        row.append(adata.obs[col].values[0])
    row.append(sub_adata.shape[1])
    row.append(sub_adata.shape[0])
    rows.append(row)
```
